# import data

```python
mono = pd.read_excel('Percent_Monocytes.xlsx')
mono
```

```python
neut = pd.read_excel('Percent_Neutrophils.xlsx')
neut
```

```python
cell = pd.merge(mono, neut, left_on='Patient_ID', right_on='Patient_ID', how='outer')
cell
```

```python
neut['Patient_ID'] = neut['Patient_ID'].str.replace('_10000', '_1000')
neut
```

```python
cell = pd.merge(mono, neut, left_on='Patient_ID', right_on='Patient_ID', how='outer')
cell
```

# difference between matplotlib and seaborn

```python
fig, ax = plt.subplots()
ax.scatter(cell['Monocyte'], cell['Neutrophils'], c='black')
ax.set_xlabel('Monocyte')
ax.set_ylabel('Neutrophils')
```

```python
female = cell[cell['Sex'] == 'Female']
male = cell[cell['Sex'] == 'Male']
```

```python
fig, ax = plt.subplots(dpi=150, figsize=(2, 2))
ax.scatter(female['Monocyte'], female['Neutrophils'], c='red', label='Female')
ax.scatter(male['Monocyte'], male['Neutrophils'], c='blue', label='Male')
ax.legend(fontsize=10, bbox_to_anchor=(1, 1))

ax.set_xlabel('Monocyte')
ax.set_ylabel('Neutrophils')
```

```python
fig, ax = plt.subplots(dpi=300, figsize=(2, 2))
sns.scatterplot(data=cell, x='Monocyte', y='Neutrophils', hue='Sex', ax=ax,palette='husl')
ax.legend(fontsize=10, bbox_to_anchor=(1, 1))
ax.set_xlabel('Monocytes')
```

```python
fig, axs = plt.subplots(1, 2)
grid = plt.GridSpec(1, 3)

axs[0] = plt.subplot(grid[:1, :2])
axs[1] = plt.subplot(grid[:1, 2:3])

plt.subplots_adjust(wspace=0.5)
sns.scatterplot(data=cell, x='Monocyte', y='Neutrophils', hue='Sex', ax=axs[0], palette='husl')
sns.violinplot(data=cell, x='Sex', y='T cell', ax=axs[1], inner=None)
sns.swarmplot(data=cell, x='Sex', y='T cell', ax=axs[1], color='black')
```
